# 안전 EDA: 5대 범죄 + 출동 데이터
- 월별 출동 추이
- 범죄-출동 상관 분석
- 연도별 점수(25점)
- 3개년 평균 점수(25점)

주의: 본 점수는 **낮을수록 안전(높을수록 스트레스/위험)** 기준입니다.


```python
from pathlib import Path
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', 200)
pd.set_option('display.width', 220)

sns.set_theme(style='whitegrid', context='talk')
plt.rcParams.update({
    'font.family': 'sans-serif',
    'font.sans-serif': ['Malgun Gothic', 'AppleGothic', 'NanumGothic', 'DejaVu Sans'],
    'axes.unicode_minus': False,
})

base = Path('safety_analystic_data')
crimes_path = base / '_5_crimes__202604231451.csv'
dispatch_path = base / 'dispatch_integrated_202604231451.csv'

crimes_raw = pd.read_csv(crimes_path, encoding='utf-8')
dispatch_raw = pd.read_csv(dispatch_path, encoding='utf-8')

print('crimes_raw shape:', crimes_raw.shape)
print('dispatch_raw shape:', dispatch_raw.shape)
```

```python
# 1) 범죄 데이터 전처리(합계 포함)
crimes = crimes_raw.copy().iloc[:, :7]
crimes.columns = ['자치구', '검거_2022', '검거_2023', '검거_2024', '발생_2022', '발생_2023', '발생_2024']

for c in ['검거_2022', '검거_2023', '검거_2024', '발생_2022', '발생_2023', '발생_2024']:
    crimes[c] = pd.to_numeric(crimes[c], errors='coerce')

for y in [2022, 2023, 2024]:
    crimes[f'검거율_{y}'] = crimes[f'검거_{y}'] / crimes[f'발생_{y}']

display(crimes.head())
```

```python
# 2) 출동 데이터 전처리(wide -> long)
dispatch = dispatch_raw.copy()
dispatch = dispatch.rename(columns={dispatch.columns[0]: 'gu_code', dispatch.columns[1]: '자치구'})

value_cols = [c for c in dispatch.columns if c not in ['gu_code', '자치구']]
dispatch_long = dispatch.melt(id_vars=['gu_code', '자치구'], value_vars=value_cols, var_name='ym_metric', value_name='value')
dispatch_long['value'] = pd.to_numeric(dispatch_long['value'], errors='coerce')

parsed = dispatch_long['ym_metric'].str.extract(r'^(\d{4})_(ed|fd|rp)$')
dispatch_long['yymm'] = parsed[0]
dispatch_long['metric'] = parsed[1]
dispatch_long = dispatch_long.dropna(subset=['yymm', 'metric']).copy()
dispatch_long['year'] = dispatch_long['yymm'].str[:2].astype(int) + 2000
dispatch_long['month'] = dispatch_long['yymm'].str[2:].astype(int)
dispatch_long['date'] = pd.to_datetime(dispatch_long['year'].astype(str) + '-' + dispatch_long['month'].astype(str).str.zfill(2) + '-01')
dispatch_long['지표'] = dispatch_long['metric'].map({'ed': '응급출동', 'fd': '화재출동', 'rp': '구조출동'})

display(dispatch_long.head())
```

```python
# 3) 월별 출동 추이
monthly_city = (
    dispatch_long.dropna(subset=['value'])
    .groupby(['date', '지표'], as_index=False)['value']
    .sum()
)

plt.figure(figsize=(16, 6))
sns.lineplot(data=monthly_city, x='date', y='value', hue='지표', marker='o')
plt.title('월별 출동 지표 추이 (서울 전체)')
plt.xlabel('월')
plt.ylabel('출동 건수')
plt.tight_layout()
plt.show()
```

```python
# 4) 연도 단위 결합 + 상관 분석
dispatch_yearly = (
    dispatch_long.dropna(subset=['value'])
    .groupby(['자치구', 'year', 'metric'], as_index=False)['value']
    .sum()
    .pivot(index=['자치구', 'year'], columns='metric', values='value')
    .reset_index()
    .rename_axis(None, axis=1)
    .rename(columns={'ed': '응급출동_연합계', 'fd': '화재출동_연합계', 'rp': '구조출동_연합계'})
)

rows = []
for _, r in crimes.iterrows():
    for y in [2022, 2023, 2024]:
        rows.append({
            '자치구': r['자치구'],
            'year': y,
            '범죄발생': r[f'발생_{y}'],
            '범죄검거': r[f'검거_{y}'],
            '검거율': r[f'검거율_{y}']
        })

crimes_long = pd.DataFrame(rows)
merged = crimes_long.merge(dispatch_yearly, on=['자치구', 'year'], how='left')

corr_cols = ['범죄발생', '범죄검거', '검거율', '응급출동_연합계', '화재출동_연합계', '구조출동_연합계']
corr_df = merged[corr_cols].corr()

plt.figure(figsize=(9, 7))
sns.heatmap(corr_df, annot=True, fmt='.2f', cmap='RdBu_r', center=0)
plt.title('범죄-출동 상관 분석')
plt.tight_layout()
plt.show()

display(corr_df)
```

```python
# 5) 연도별 점수 분리(치안/재난, 각 25점)
# 정규화는 전체기간(2022~2024) 고정 Z-score 기준
# 점수는 모두 낮을수록 안전
score_base = merged.copy()

# 컬럼 탐색(한글 인코딩 영향 최소화)
district_col = merged.columns[0]
year_col = 'year'
crime_col = [c for c in merged.columns if '범죄발생' in c][0]
arrest_rate_col = [c for c in merged.columns if '검거율' in c][0]
dispatch_cols = [c for c in merged.columns if c.endswith('_연합계')]

score_base = score_base[~score_base[district_col].isin(['소계', '합계'])].copy()
score_base['dispatch_total'] = score_base[dispatch_cols].sum(axis=1, skipna=True)

# 전체기간 고정 Z-score 함수
# std가 0이면 0 반환(변별 불가)
def zscore_fixed(s):
    s = s.astype(float)
    std = s.std(ddof=0)
    if pd.isna(std) or std == 0:
        return pd.Series(0.0, index=s.index)
    return (s - s.mean()) / std

# z-score를 점수용 0~1 구간으로 변환(이상치 과민 방지)
def z_to_unit(z, clip=3.0):
    z = z.clip(-clip, clip)
    return (z + clip) / (2 * clip)

# 전체기간 기준 z-score(고정 기준)
score_base['z_crime'] = zscore_fixed(score_base[crime_col])
score_base['z_arrest'] = zscore_fixed(score_base[arrest_rate_col])
score_base['z_disaster'] = zscore_fixed(score_base['dispatch_total'])

# 방향 정렬: 값이 클수록 위험(나쁨)
score_base['crime_risk_norm'] = z_to_unit(score_base['z_crime'])
score_base['arrest_lack_norm'] = z_to_unit(-score_base['z_arrest'])
score_base['disaster_load_norm'] = z_to_unit(score_base['z_disaster'])

# 25점 점수 산출
score_base['security_score_25'] = (
    15 * score_base['crime_risk_norm']
    + 10 * score_base['arrest_lack_norm']
).round(2)

score_base['disaster_score_25'] = (25 * score_base['disaster_load_norm']).round(2)

# 연도별 순위 계산(점수는 고정 기준, 순위만 연도 내 산출)
score_parts = []
for y in sorted(score_base[year_col].dropna().unique()):
    d = score_base[score_base[year_col] == y].copy()
    d['security_rank_year'] = d['security_score_25'].rank(method='dense', ascending=True).astype(int)
    d['disaster_rank_year'] = d['disaster_score_25'].rank(method='dense', ascending=True).astype(int)
    score_parts.append(d)

score_yearly_split = pd.concat(score_parts, ignore_index=True)

display(
    score_yearly_split[[
        year_col, district_col,
        'security_score_25', 'security_rank_year',
        'disaster_score_25', 'disaster_rank_year',
        crime_col, arrest_rate_col, 'dispatch_total'
    ]].sort_values([year_col, 'security_rank_year', 'disaster_rank_year'])
)

score_3y_split = (
    score_yearly_split.groupby(district_col, as_index=False)
    .agg(
        security_score_3y_avg=('security_score_25', 'mean'),
        disaster_score_3y_avg=('disaster_score_25', 'mean'),
        crime_3y_avg=(crime_col, 'mean'),
        arrest_rate_3y_avg=(arrest_rate_col, 'mean'),
        dispatch_total_3y_avg=('dispatch_total', 'mean')
    )
)

score_3y_split['security_score_3y_avg'] = score_3y_split['security_score_3y_avg'].round(2)
score_3y_split['disaster_score_3y_avg'] = score_3y_split['disaster_score_3y_avg'].round(2)
score_3y_split['security_rank_3y'] = score_3y_split['security_score_3y_avg'].rank(method='dense', ascending=True).astype(int)
score_3y_split['disaster_rank_3y'] = score_3y_split['disaster_score_3y_avg'].rank(method='dense', ascending=True).astype(int)

display(
    score_3y_split[[
        district_col,
        'security_score_3y_avg', 'security_rank_3y',
        'disaster_score_3y_avg', 'disaster_rank_3y',
        'crime_3y_avg', 'arrest_rate_3y_avg', 'dispatch_total_3y_avg'
    ]].sort_values(['security_rank_3y', 'disaster_rank_3y'])
)

# 시각화: 3개년 평균 점수(낮을수록 안전)
fig, axes = plt.subplots(1, 2, figsize=(18, 8), sharey=True)

plot_sec = score_3y_split.sort_values('security_score_3y_avg', ascending=True)
sns.barplot(data=plot_sec, x='security_score_3y_avg', y=district_col, ax=axes[0], palette='Blues_r')
axes[0].set_title('자치구별 3개년 평균 치안점수 (낮을수록 안전)')
axes[0].set_xlabel('security_score_3y_avg')
axes[0].set_ylabel('자치구')

plot_dis = score_3y_split.sort_values('disaster_score_3y_avg', ascending=True)
sns.barplot(data=plot_dis, x='disaster_score_3y_avg', y=district_col, ax=axes[1], palette='Oranges_r')
axes[1].set_title('자치구별 3개년 평균 재난점수 (낮을수록 안전)')
axes[1].set_xlabel('disaster_score_3y_avg')
axes[1].set_ylabel('')

plt.tight_layout()
plt.show()
```

```python
# 6) 결과 저장
out_dir = Path('safety_analystic_data')
merged.to_csv(out_dir / 'safety_crimes_dispatch_merged.csv', index=False, encoding='utf-8-sig')
score_yearly_split.to_csv(out_dir / 'safety_score_yearly_split_25point.csv', index=False, encoding='utf-8-sig')
score_3y_split.to_csv(out_dir / 'safety_score_3year_avg_split_25point.csv', index=False, encoding='utf-8-sig')
print('saved:', out_dir / 'safety_crimes_dispatch_merged.csv')
print('saved:', out_dir / 'safety_score_yearly_split_25point.csv')
print('saved:', out_dir / 'safety_score_3year_avg_split_25point.csv')
```

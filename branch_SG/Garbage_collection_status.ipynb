```python
import os
import glob
import gc
import pandas as pd
```

```python
path = r'C:\8ssible-Healing-Seoul-Analysis\branch_SG\8SSIBLE_SG_DATA\쓰레기수거+현황(2020년+이후)_20260422193426.csv'

df_raw = pd.read_csv(path, encoding='utf-8-sig')

# 첫 2행은 헤더 정보, 2행부터 실제 데이터
df_trash = df_raw.iloc[2:].copy()

# 컬럼명 만들기
new_cols = ['구분1', '자치구']

for col, upper, lower in zip(df_raw.columns[2:], df_raw.iloc[0, 2:], df_raw.iloc[1, 2:]):
    year = col.split('.')[0]
    new_cols.append(f'{year}_{upper}_{lower}')

df_trash.columns = new_cols

# 인덱스 정리
df_trash = df_trash.reset_index(drop=True)

df_trash.head()
```

```python
df_trash = df_trash.drop(columns = '구분1')
```
